```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from linear_classifiers import GDA, LogisticRegression
```

```python
ds1_train = pd.read_csv("data/ds1_train.csv")
ds1_valid = pd.read_csv("data/ds1_valid.csv")
model1 = LogisticRegression()
model2 = GDA()
```

```python
ds1_train_features = np.array(ds1_train)[:,0:2]
ds1_train_labels = np.array(ds1_train)[:,2]
ds1_valid_features = np.array(ds1_valid)[:,0:2]
ds1_valid_labels = np.array(ds1_valid)[:,2]
```

```python
#let's plot the data
ds1_train_x1 = ds1_train_features[:,0]
ds1_train_x2 = ds1_train_features[:,1]
ds1_train_y = ds1_train_labels
plt.scatter(ds1_train_x1[ds1_train_y == 1], ds1_train_x2[ds1_train_y == 1], alpha= 0.3, label='y = 1')
plt.scatter(ds1_train_x1[ds1_train_y == 0], ds1_train_x2[ds1_train_y == 0], alpha= 0.3, label='y = 0')
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.legend()
plt.show()
```

```python
model1.train_model(ds1_train_features, ds1_train_labels, theta_0=[0,0,0])
model2.train_model(ds1_train_features, ds1_train_labels)
print("trained the models")
```

```python
# now let's plot the model results!!
x, y = np.meshgrid(np.linspace(ds1_train_x1.min(), ds1_train_x1.max(), 100), np.linspace(ds1_train_x2.min(), ds1_train_x2.max(), 100))
plt.contour(x, y, model1.theta[0] + model1.theta[1]*x + model1.theta[2]*y, levels=[0], colors = 'red')
plt.contour(x, y, model2.theta[0] + model2.theta[1]*x + model2.theta[2]*y, levels=[0], colors = 'purple')
plt.scatter(ds1_train_x1[ds1_train_y == 1], ds1_train_x2[ds1_train_y == 1], alpha= 0.3, label='y = 1')
plt.scatter(ds1_train_x1[ds1_train_y == 0], ds1_train_x2[ds1_train_y == 0], alpha= 0.3, label='y = 0')
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.legend()
plt.show()
```

```python
#now let's review on the test data:
ds1_test_x1 = ds1_valid_features[:,0]
ds1_test_x2 = ds1_valid_features[:,1]
ds1_test_y = ds1_valid_labels
x, y = np.meshgrid(np.linspace(ds1_test_x1.min(), ds1_test_x1.max(), 100), np.linspace(ds1_test_x2.min(), ds1_test_x2.max(), 100))
plt.contour(x, y, model1.theta[0] + model1.theta[1]*x + model1.theta[2]*y, levels=[0], colors = 'red')
plt.contour(x, y, model2.theta[0] + model2.theta[1]*x + model2.theta[2]*y, levels=[0], colors = 'purple')
plt.scatter(ds1_test_x1[ds1_test_y == 1], ds1_test_x2[ds1_test_y == 1], alpha= 0.5, label='y = 1')
plt.scatter(ds1_test_x1[ds1_test_y == 0], ds1_test_x2[ds1_test_y == 0], alpha= 0.5, label='y = 0')
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.legend()
plt.show()
```

```python
y_pred_ds1_logreg = model1.make_prediction(ds1_valid_features)
y_pred_ds1_gda = model2.make_prediction(ds1_valid_features)
accuracy_ds1_logreg = 1 - 1/len(ds1_test_y) * np.sum(np.abs(y_pred_ds1_logreg - ds1_test_y))
accuracy_ds1_gda = 1 - 1/len(ds1_test_y) * np.sum(np.abs(y_pred_ds1_gda - ds1_test_y))
```

```python
print("Logistic reg accuracy: ", accuracy_ds1_logreg)
print("GDA accuracy: ", accuracy_ds1_gda)
```
