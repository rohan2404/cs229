```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
ds1_a = pd.read_csv("data/ds1_a.csv")
ds1_b = pd.read_csv("data/ds1_b.csv")
```

```python
y_a = ds1_a["y"]
X_a = np.column_stack([ds1_a["x0"], ds1_a["x1"]])
X_train_a, X_test_a, y_train_a, y_test_a = train_test_split(X_a, y_a, test_size=0.2)
```

```python
plt.scatter(X_train_a[:,0][y_train_a==1],X_train_a[:,1][y_train_a==1], color = 'red', linewidths=0.1)
plt.scatter(X_train_a[:,0][y_train_a==-1],X_train_a[:,1][y_train_a==-1], color = 'blue', linewidths=0.1)
plt.show()
```

```python
model1 = LogisticRegression(
    penalty='l2', max_iter=1000,
)
model1.fit(X_train_a, y_train_a)
```

```python
y_test_pred_a = model1.predict(X_test_a) #this test is technically a valid set since i will be drawing conclusions from this
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("Accuracy:", accuracy_score(y_test_a, y_test_pred_a))
print("\nClassification Report:\n", classification_report(y_test_a, y_test_pred_a))
print("Confusion Matrix:\n", confusion_matrix(y_test_a, y_test_pred_a))
```
