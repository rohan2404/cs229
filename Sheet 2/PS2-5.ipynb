```python
import numpy as np
import pandas as pd
import perceptron as pp
import matplotlib.pyplot as plt
```

```python
def load_data(dataframe: pd.DataFrame):
    return np.array(dataframe[dataframe.columns[0]]),np.array(dataframe[dataframe.columns[1]]),np.array(dataframe[dataframe.columns[2]])
y_train, x0_train, x1_train = load_data(pd.read_csv('data/ds5_train.csv'))
X = np.column_stack([x0_train, x1_train])
```

```python
model = pp.Gaussian_Kernel()
```

```python
plt.plot(x0_train[y_train == 1], x1_train[y_train == 1], '.', color ='red')
plt.plot(x0_train[y_train == 0], x1_train[y_train == 0], '.', color ='blue')
plt.show()
```

```python
model.trainSGD(y_train, X, lr = 0.1, gamma = 0.1, iter=1000)
z = model.make_prediction(X)
```

```python
plt.plot(x0_train[z == 1], x1_train[z == 1], '.', color ='red')
plt.plot(x0_train[z == 0], x1_train[z == 0], '.', color ='blue')
plt.show()
```

```python
#looks good (cba to put accuracy measures)- let's test it!
y_test, x0_test, x1_test = load_data(pd.read_csv('data/ds5_test.csv'))
X = np.column_stack([x0_test, x1_test])
```

```python
z = model.make_prediction(X)
```

```python
w = z - y_test
correct = len(w[w==0])
print(f"proportion of correct classifications is {correct/len(w)}")
```
